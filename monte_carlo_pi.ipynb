Estimating the value of $\pi$ using Monte Carlo methods

```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.patches as patches
```

The core idea is to 

1. Generate random points in the region defined by $(-1,-1), (-1,1),(1,-1)$ and $(1,1)$

2. Count how many points are in the circle with center $(0,0)$

3. The value of $\pi$ is given by 4* number of random points inside the circle/total number of random points

```python
points = 10000
#generating random numbers between -1 and 1
rand = 2* np.random.rand(2*points) - 1

#reshape the array so we have points like (x,y)
randpoints = rand.reshape(points, 2)

```

```python
#calculate the norm of the points
normpoints = randpoints[:,0]**2 + randpoints[:,1]**2
```

```python
# this is the cool bit, this technique gets all the points which 
#satisfy the condition inside the square brackets

pointsOut = randpoints[normpoints > 1]
pointsIn = randpoints[normpoints<=1]
```

```python
piApprox = 4 * len(pointsIn)/points
```

Finally, we see that with 10k random points, the difference between the calculated value and the "real" value is off by 0.003

```python
piApprox - np.pi
```

Visualization-
